```python
import numpy as np
import pandas as pd
from math import exp
from tqdm import tqdm_notebook as tqdm
```

```python
data = pd.read_csv('../all.csv')
corrs = pd.read_csv('../cors.csv')
```

```python
mu = data.groupby('Name').apply(lambda x : x['Close'].pct_change(22).dropna().mean())
sigma = data.groupby('Name').apply(lambda x : x['Close'].pct_change(22).dropna().std())

mu = pd.DataFrame({ 'name': mu.index, 'value': mu }).sort_values(['name']).reset_index(drop=True)
sigma = pd.DataFrame({ 'name': sigma.index, 'value': sigma }).sort_values(['name']).reset_index(drop=True)
mu['value'].fillna(0, inplace=True)
sigma['value'].fillna(0, inplace=True)

mu_dict = { mu.loc[i, 'name']: mu.loc[i, 'value'] for i in list(mu.index) }
sigma_dict = { sigma.loc[i, 'name']: sigma.loc[i, 'value'] for i in list(sigma.index) }
```

```python
name_to_index = { mu.loc[i, 'name']: i for i in list(mu.index) }
```

```python
corrs = corrs.sort_values(['name1', 'name2'])[['name1', 'name2', 'pearson_year_cor']]
```

```python
corrs['index1'] = corrs['name1'].apply(lambda x : name_to_index[x])
corrs['index2'] = corrs['name2'].apply(lambda x : name_to_index[x])
```

```python
corrs['cov'] = corrs.apply(lambda row : row['pearson_year_cor'] * sigma_dict[row['name1']] * sigma_dict[row['name2']], axis=1)
corrs['cov'].fillna(0, inplace=True)
```

```python
C = np.zeros((len(mu), len(mu)))
```

```python
for i, row in tqdm(corrs.iterrows()):
    C[row['index1'], row['index2']] = row['cov']
    C[row['index2'], row['index1']] = row['cov']
for i in range(len(mu)):
    C[i, i] = sigma.loc[i, 'value'] ** 2
```

```python
mu = np.array(mu.value)
```

```python
L = lambda w, mu, C : gamma * np.matmul(mu, w.transpose()) - (1 - gamma) * np.matmul(w, np.matmul(C, w.transpose()))
```

```python
gamma = 0.99
gamma_u = 0.99
u = 0.1
T = 100
K = 10
N = 443
I = int(N * (N - 1) / 2)

w = np.zeros(N)
fixed_sum = 1
choices = np.random.choice(N, K, replace=False)
for i in range(len(choices) - 1):
    value = np.random.uniform(0, fixed_sum)
    w[choices[i]] = value
    fixed_sum -= value
w[choices[-1]] = fixed_sum

ws = w
for i in tqdm(range(I)):
    wu = w
    choices = np.where(w != 0.0)[0]
    i = np.random.choice(choices)
    j = np.random.randint(N)
    while j != i:
        j = np.random.randint(N)
    if w[j] == 0:
        wu[i] = 0
        wu[j] = w[i]
    else:
        ru = np.random.uniform(0, u)
        wu[i] = w[i] - ru
        wu[j] = w[j] + ru
    delta_L = L(wu, mu, C) - L(w, mu, C)
    if delta_L > 0:
        w = wu
        if L(w, mu, C) > L(ws, mu, C):
            ws = w
    else:
        p = np.random.uniform(0, 1)
        if p < exp(delta_L / T):
            w = wu
    T *= gamma
    u = gamma_u * u
```
